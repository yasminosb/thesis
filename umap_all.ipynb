```python
import pandas as pd
import json

from ipynb.fs.full.common import *
```

```python
# load gameplays
gameplays = pd.json_normalize(pd.Series(open(r"C:\Users\Yasmine Bogaert\Documents\thesis\dump\dump_30-03-2021\csv\gameplays.json").readlines()).apply(json.loads))
gameplays = preprocess_gameplays(gameplays)
gameplays = add_distance_per_jump(gameplays)
```

```python
# load questionresponses and expand the response columns to contain "lastentry" and "secondlastentry" instead of 'A', 'B' or 'BOTH'
questionresponses = pd.json_normalize(pd.Series(open(r"C:\Users\Yasmine Bogaert\Documents\thesis\dump\dump_30-03-2021\csv\questionresponses.json").readlines()).apply(json.loads))
questionresponses = expand_questionresponses(questionresponses)
```

```python
import umap
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import math

sns.set(style='white', context='notebook', rc={'figure.figsize':(30,20)})

a = explode_questionresponses(questionresponses, gameplays)
def draw_umap_parameter(parameter, n_neighbors=15, min_dist=0.1,  metric='euclidean', nr_plots = 1):
    game_data = a[[
    'fun',
    'challenging',
    'frustrating',
    parameter
    ]].values
    scaled_data = StandardScaler().fit_transform(game_data)
    
    fit = umap.UMAP(
        n_neighbors=n_neighbors,
        min_dist=min_dist,
        metric=metric
    )
    fig = plt.figure()
    sq = int(math.ceil(math.sqrt(nr_plots)))
    nr = sq * 100 + sq * 10
    for i in range(nr_plots):
        ax = fig.add_subplot(nr + 1 + i)
        u = fit.fit_transform(scaled_data);
        p = ax.scatter(u[:,0], u[:,1], c=a[parameter], cmap=sns.color_palette("rocket_r", as_cmap=True))
        fig.colorbar(p)
        plt.title(parameter, fontsize=18)

draw_umap_parameter('actualDistance', nr_plots = 2)
plt.savefig('graphs/notebook/umap/umap_evaluations_actualDistance_2.png')
```

```python
draw_umap_parameter('parameters.SPEED', nr_plots = 2)
plt.savefig('graphs/notebook/umap/umap_evaluations_speed_2.png')
```

```python
# converst collisionObstacle to ints
dict = {'CACTUS_LARGE':0, 'CACTUS_SMALL':1, 'PTERODACTYL':2}
if(isinstance(a['collisionObstacle.typeConfig.type'][0], str)):
    a['collisionObstacle.typeConfig.type'] = a['collisionObstacle.typeConfig.type'].replace(dict)


def draw_umap_all_highlight_parameter(parameter, n_neighbors=15, min_dist=0.1,  metric='euclidean', nr_plots = 1):
    game_data = a[[
          'fun',
          'challenging',
          'frustrating',
          'actualDistance',
          'parameters.SPEED', 
          'parameters.ACCELERATION',
          'parameters.MIN_GAP',
          'parameters.NIGHT_MODE_ENABLED', 
          'parameters.NIGHT_MODE_DISTANCE',
          'parameters.CLEAR_TIME', 
          'parameters.MAX_OBSTACLE_LENGTH',
          'parameters.MAX_SPEED',
          'parameters.MAX_GAP', 
          'actualDistance', 
          'collisionObstacle.typeConfig.type',
          'nr_jumps',
          'invertedGameOver'
    ]].values
    scaled_data = StandardScaler().fit_transform(game_data)
    
    fit = umap.UMAP(
        n_neighbors=n_neighbors,
        min_dist=min_dist,
        metric=metric
    )
    fig = plt.figure()
    sq = int(math.ceil(math.sqrt(nr_plots)))
    nr = sq * 100 + sq * 10
    for i in range(nr_plots):
        ax = fig.add_subplot(nr + 1 + i)
        u = fit.fit_transform(scaled_data);
        p = ax.scatter(u[:,0], u[:,1], c=a[parameter], cmap=sns.color_palette("rocket_r", as_cmap=True))
        ax.text(0.99, 0.99, "n_neighbors: " + str(n_neighbors) + ", min_dist: " + str(min_dist)  , horizontalalignment='right', verticalalignment='top',  transform=ax.transAxes)
        fig.colorbar(p)
        plt.title(parameter, fontsize=18)


sns.set(style='white', context='notebook', rc={'figure.figsize':(15,10)})
draw_umap_all_highlight_parameter('actualDistance')
plt.savefig('graphs/notebook/umap/umap_evaluations_everything.png')
```

```python
def draw_umap_all_highlight_parameter_(parameter, ax=None , n_neighbors=15, min_dist=0.1,  metric='euclidean'):
    game_data = a[[
          'fun',
          'challenging',
          'frustrating',
          'actualDistance',
          'parameters.SPEED', 
          'parameters.ACCELERATION',
          'parameters.MIN_GAP',
          'parameters.NIGHT_MODE_ENABLED', 
          'parameters.NIGHT_MODE_DISTANCE',
          'parameters.CLEAR_TIME', 
          'parameters.MAX_OBSTACLE_LENGTH',
          'parameters.MAX_SPEED',
          'parameters.MAX_GAP', 
          'actualDistance', 
          'collisionObstacle.typeConfig.type',
          'nr_jumps',
          'invertedGameOver'
    ]].values
    scaled_data = StandardScaler().fit_transform(game_data)
    
    fit = umap.UMAP(
        n_neighbors=n_neighbors,
        min_dist=min_dist,
        metric=metric
    )
    u = fit.fit_transform(scaled_data);
    p = ax.scatter(u[:,0], u[:,1], c=a[parameter], cmap=sns.color_palette("rocket_r", as_cmap=True))
    ax.text(0.99, 0.99, "n_neighbors: " + str(n_neighbors) + ", min_dist: " + str(min_dist)  , horizontalalignment='right', verticalalignment='top',  transform=ax.transAxes)
    fig.colorbar(p, ax=ax)
    ax.title.set_text(parameter)


sns.set(style='white', context='notebook', rc={'figure.figsize':(30,20)})

ns = [2, 5, 10, 20, 50, 100, 200]
ds = [0.0, 0.1, 0.25, 0.5, 0.8, 0.99]
x = len(ns)
y = len(ds)

fig, ax_array = plt.subplots(x, y)
for i, n in enumerate(ns):
    for j, d in enumerate(ds):
        print( "({},{}) from ({},{})".format(i,j,len(ns),len(ds)) )
        ax = ax_array[i,j]
        draw_umap_all_highlight_parameter_("actualDistance", ax=ax, n_neighbors=n, min_dist=d)

plt.savefig('graphs/notebook/umap/umap_evaluations_everything_iterations.png')
```
